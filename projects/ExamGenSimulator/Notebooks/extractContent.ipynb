```python
import json
import requests
from bs4 import BeautifulSoup

api_key = '551ef1f8c67e4dc3ab1f7cb13d6578f9'
course_url = 'https://learn.microsoft.com/en-us/training/modules/introduction-end-analytics-use-microsoft-fabric/2-explore-analytics-fabric'

endpoint = 'https://api.scrapingant.com/v2/general'
headers = {'x-api-key': api_key}

params = {
    'url': course_url,
    'x-api-key': api_key,
    'browser': False
}

# Get course content
response = requests.get(endpoint, params=params, headers=headers)
if response.status_code != 200:
    print(f"Error while scraping the course: {response.status_code}")
    # return {}

soup = BeautifulSoup(response.text, 'html.parser')

div = soup.find("div", id="module-unit-content")
if div:
    print(div.get_text(separator=" ", strip=True))
else:
    print(f"<div id='module-unit-content'> not found on {course_url}")
    # return None
```

```python
from langchain.document_transformers.html2text import Html2TextTransformer
from langchain.schema import Document

#HTML from BeautifulSoup
html_content = str(soup)

# Create a Document object with the HTML as “page_content”
doc = Document(page_content=html_content)

# Apply the HTML → text transformer
transformer = Html2TextTransformer() #Markdown Characters
#transformer = Html2TextTransformer(html2text_kwargs={"bodywidth": 0, "ignore_links": True, "ignore_emphasis": True})

docs_transformed = transformer.transform_documents([doc])

# docs_transformed is a list of transformed documents
for d in docs_transformed:
    print("Transformed text:")
    print(d.page_content)
```
